## 1.1) Splitting the data into Train and Test and completing missing data


* #### Since we are dealing with a prediction which supposes to be in the future, then the right splitting will be to take the latest data collected as our test data and the rest will be train data. I will use November as our test data (since it is our last month in our data set)!

```python
# orgenize the data from old to new 
df = df.sort_values(["month", "day","hour"], ascending = (True,True, True))
bollean_vec = df['month'] == 11
oposite = [not i for i in bollean_vec]
df_test = df[bollean_vec]
df_train = df[oposite]
```

* #### First, let's fix user_isp feature by using sampling! 

* #### Since user_isp strongly depends on the user state, we will do sampling over users with the same state (note this feature is not missing in the test set, so we don't need to complete it in the test set)

```python
boolean_condition = df_train['user_isp'].isnull() 
df_sub = df_train[boolean_condition]
boolean_condition_fliped = [not i for i in boolean_condition]
for i in df_sub.index:
    
    user_state_i = df_sub.loc[i].user_state
    boolean_condition2 = df_train['user_state'] == user_state_i 
    df_train_local = df_train[boolean_condition2 & boolean_condition_fliped]
    v =  df_train_local.user_isp.sample(n = 1, replace=True)
    df_train.loc[i,'user_isp'] = v.values
    
df_train.info(null_counts=True)
```


* #### Now, let's fix the device_maker feature! 

* #### I assume that the device_maker is dependent on the user_state (because most of the time the stores in the same country have the same variety of phones). Here we need to fix also in the test set!

```python
from tqdm import tqdm

boolean_condition = df_train['device_maker'].isnull() 
df_sub = df_train[boolean_condition]

boolean_condition_fliped = [not i for i in boolean_condition]

states = df_sub.user_state.unique()

boolean_condition_test = df_test['device_maker'].isnull()

for s in tqdm(states):
    user_state_i = s
    boolean_condition2 = df_train['user_state'] == user_state_i 
    boolean_condition2_test = df_test['user_state'] == user_state_i
    
    df_train_local = df_train[boolean_condition2 & boolean_condition_fliped]
    
    bool_to_fix = boolean_condition & boolean_condition2
    v =  df_train_local.device_maker.sample(n = sum(bool_to_fix), replace=True)
    
    
    bool_to_fix_test = boolean_condition_test & boolean_condition2_test
    v_test =  df_train_local.device_maker.sample(n = sum(bool_to_fix_test), replace=True)
    df_train.loc[bool_to_fix,'device_maker'] = v.values
    df_test.loc[bool_to_fix_test, 'device_maker'] = v_test.values
    
```

* #### As we can see there are more users in some states than other states! Let's see what happen if we do the same using the area of the state ('West South, East South, East North, .... )

```python

state_to_region = {'WS': ['OK','AR', 'LA','TX'],
                   'ES': ['KY','TN', 'WV', 'VA', 'DC', 'MD',
                          'DE','MS', 'AL', 'GA', 'FL', 'SC', 
                          'NC'],
                   'EN': ['PA', 'ME', 'NH', 'VT', 'MA', 
                          'CT', 'RI', 'NJ', 'NY'],
                   'ENC': ['WI', 'MI', 'OH', 'IN', 'IL'],
                   'WNC': ['ND', 'SD', 'NE', 'KS', 'MO',
                          'MN', 'IA'],
                   'W': ['WA', 'OR', 'CA', 'NV', 
                         'AZ', 'NM','CO','UT',
                         'WY', 'ID','MT', 'AK', 'HI']}

def find_region(s):
    global state_to_region
    for re in state_to_region.keys():
        if s in state_to_region[re]:
            return re
    
    #print(s)
    return 


regions = []
states = df_train.user_state.unique()
df_train['region'] = None

for s in states:
    
    b = df_train['user_state'] == s
    
    region = find_region(s)
    
    df_train.loc[b, 'region'] = region
    
```

# 4) Hyperparam tunning based on optuna (Bayesian approach)

```python
from imblearn.over_sampling import SMOTE
df_train  = original_train
df_test = original_test

X_train = df_train.drop(['click'], axis = 1).to_numpy()
Y_train = df_train['click'].to_numpy()

X_test = df_test.drop(['click'], axis = 1).to_numpy()
Y_test = df_test['click'].to_numpy()

oversample_smote = SMOTE()
X_train, Y_train = oversample_smote.fit_resample(X_train, Y_train)

X_train, Y_train = shuffle(X_train, Y_train)

n = 500000
X_train_batch = X_train[:n]
Y_train_batch = Y_train[:n]
```

```python

model = XGBClassifier()



scores = cross_val_score(model, X_train_batch, Y_train_batch, scoring='accuracy', n_jobs=-1, cv=5)
model.fit(X_train_batch, Y_train_batch)


print('Mean Accuracy: %.3f (%.3f)' % (scores.mean(), scores.std()) )

print("test score:", model.score(X_test, Y_test))
print(classification_report(Y_test, model.predict(X_test)))
cf = confusion_matrix(Y_test, model.predict(X_test)).T
d = pd.DataFrame(cf, columns=['0', '1'])
print("Confusion Matrix:")
print(d)

```

```python


import optuna

def objective(trial):
    global X_train_batch
    global Y_train_batch
    
    n_estimators = trial.suggest_int('n_estimators', 2,160)# computational expensive, so the range was limited up to 160  
    max_depth = trial.suggest_int('max_depth', 1, 10)
    # lr = trial.suggest_int('learning_rate',0.001 ,0.2)
    
    clf = XGBClassifier(n_estimators=n_estimators, max_depth=max_depth)
    
    return cross_val_score(clf, X_train_batch, Y_train_batch, n_jobs=-1, cv=3).mean()

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials= 20)

trial = study.best_trial

print(f'Accuracy: {trial.value}')
print(f'Best hyperparameters: {trial.params}')
```

```python
# The last run for analysing the whole data and statistics
import matplotlib.pyplot as plt
optimal_n_estimators = 135#135# trial.params['n_estimators'] 
optimal_max_depth = 9 #trial.params['max_depth']

model = XGBClassifier(n_estimators = optimal_n_estimators,
                      max_depth = optimal_max_depth)



#scores = cross_val_score(model, X_train, Y_train, scoring='accuracy', n_jobs=-1, cv=5)
eval_set = [(X_train, Y_train), (X_test, Y_test)]
model.fit(X_train, Y_train, eval_metric=["error", "logloss"], eval_set=eval_set, verbose=True)



#print('Mean Accuracy: %.3f (%.3f)' % (scores.mean(), scores.std()) )

print("test score:", model.score(X_test, Y_test))
print(classification_report(Y_test, model.predict(X_test)))
cf = confusion_matrix(Y_test, model.predict(X_test)).T
d = pd.DataFrame(cf, columns=['0', '1'])
print("Confusion Matrix:")
print(d)



# make predictions for test data
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]

# retrieve performance metrics
results = model.evals_result()
epochs = len(results['validation_0']['error'])
x_axis = range(0, epochs)

```
